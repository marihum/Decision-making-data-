## Importación de librerías y carga de base de datos

```python
import pandas as pd 
import numpy as np
```

```python
df= pd.read_csv('oceania_covid.csv')
df.head(10)
```

## Description of the dataset

### Esta base de datos muestra el comportamiento de la enfermedad Covid-19 en el continente de Oceanía.

```python
df.columns
```

### En este caso me enfocaré en las variables de Total cases, Total deaths,Total recovered, Country and Active cases.

```python
df.drop(['Tot Cases/ 1M pop','Deaths/ 1M pop','Total Tests','Tests/ 1M pop','Population',],
        axis= 1,
        inplace= True)
df.columns
```

```python
df.rename(columns = { 'Country/Other':'Country',},
          inplace= True
         )
df.head()
```

```python
df.columns= list(map(str,df.columns))
df.columns
```

## Descripcion de las variables

#### Country- Es una variable categórica que contiene los países estudiados en la base de datos. 
#### Total Cases- Es una variable numérica la cual representa el número total de casos de covid-19.
#### Total deaths - Es una variable numérica la cual representa el número total de muertes por covid-19.
#### Total recovered - Es una variable numérica la cual representa el número total de pacientes recuperados.
#### Active cases - Es una variable numérica la cual representa el número total de casos activos.
