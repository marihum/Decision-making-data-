```python
import pandas as pd
```

```python
df = pd.read_excel('mic2 (3).xlsx',
engine ='openpyxl')
                   #df
```

```python
def custom_cls(x):
  if (0<= x <=20):
    return '<=20'
  elif(21<= x <=30):
    return '21 - 30'
  elif (31 <= x <= 40):
    return '31 - 40'
  else:
    return '> 40'
```

```python
df ['AgeGP'] = df.apply(lambda tbl : custom_cls(tbl['Age']),
                        axis = 1)
#df.head(2)
```

```python
age_gender = df.groupby(['AgeGP','Sex']).size()
age_gender = age_gender.to_frame()
#age_gender
```

```python
age_sex = pd.pivot_table(index = ['AgeGP','Sex'],
                                  aggfunc = len,
                                  values = 'Availability',
                                  data = df,
                                  margins = True)
age_sex
```

```python
age_gender = age_gender.rename(columns = {0:'Count'})
age_gender = age_gender.reset_index()
age_gender#.columns
```

```python
x = age_gender.iloc[:,-1].sum()
x
age_gender['%'] = (age_gender.iloc[:,-1] / x ) *100
age_gender
```

```python
plot_order = ['<=20', '21 - 30', '31 - 40', '> 40']
```

```python
import seaborn as sns
```

```python
g = sns.catplot (x = 'AgeGP',
                 kind = 'count',
                 data = df,
                 hue = 'Sex',
                 order = plot_order,
                 height =5,
                 aspect = 1.5)
```

```python
import matplotlib as matplotlib
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
fig = px.sunburst(age_gender,
                  values = 'Count',
                  path = ['AgeGP','Sex','Count','%'],
                  color = 'AgeGP',
                  title = 'Number of respondents by AgeGroup an Gender'
                  )
fig.show()
```
